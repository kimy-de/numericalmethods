```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time domain
endtime = 2*np.pi
t = np.linspace(0,endtime,50)

# r(t)
x = np.cos(t)
y = np.sin(t)

# F(x,y)=(x-y)i+xj
Fx = x-y
Fy = x

M = np.sqrt((Fx)**2 + (Fy)**2) 
plt.figure(figsize=(10,8))
q = plt.quiver(x, y, Fx, Fy, M, cmap=plt.cm.jet, scale=20)
plt.plot(x,y)
plt.colorbar(q, cmap=plt.cm.jet) 
plt.xlim(np.min(x)-0.1,np.max(x)+0.1)
plt.ylim(np.min(y)-0.1,np.max(y)+0.1)
plt.show()
```

```python
# Time domain
endtime = 30
t = np.linspace(0,endtime,500)

# r(t)
x = np.cos(t)
y = np.sin(t)
z = t/(3*np.pi)

# F(x,y,z)
Fx = x-y
Fy = y*z
Fz = 0.1*np.sqrt(x**2+y**2)

M = np.sqrt((Fx)**2 + (Fy)**2 + (Fz)**2) 
fig = plt.figure(figsize=(10,8))
fig.add_subplot(projection='3d')
q = plt.quiver(x, y, z, Fx, Fy, Fz, length=0.1, colors=plt.cm.hsv(M), cmap=plt.cm.jet)
plt.plot(x,y,z)
plt.colorbar(q, cmap=plt.cm.jet) 
plt.xlim(np.min(x)-0.1,np.max(x)+0.1)
plt.ylim(np.min(y)-0.1,np.max(y)+0.1)
plt.show()
```

```python
# Spatial domain
x, y = np.meshgrid(np.linspace(0,1,25), np.linspace(0,1,25))

# F(x,y,z)
Fx = np.cos(x)
Fy = np.sin(y)

M = np.sqrt((Fx)**2 + (Fy)**2) 
plt.figure(figsize=(10,8))
q = plt.quiver(x, y, Fx, Fy, M, cmap=plt.cm.jet)
plt.colorbar(q, cmap=plt.cm.jet) 
plt.xlim(np.min(x)-0.1,np.max(x)+0.1)
plt.ylim(np.min(y)-0.1,np.max(y)+0.1)
plt.show()
```
